```python
import pandas as pd
import numpy as np
import pylab as pl
```

```python
# url = "https://bit.ly/videosfile"
url = "https://raw.githubusercontent.com/fedhere/FDSfE_FBianco/main/data/USyoutubes.csv"

videos = pd.read_csv(url)
videos
```

```python
videos.describe()
# Qui mi restituisce solo le colonne numeriche (non ha senso calcolare media 
# di stringhe)
```

```python
type(videos.describe())
```

```python
ncols = videos.describe().columns
# Usiamo per trovare i nomi delle colonne numeriche
```

```python
videos[ncols].mean()
```

```python
videos[ncols]
```

```python
# Per trasformare il contenuto di un dataframe in un array
videos_array = videos[ncols].values
videos_array.shape
```

```python
videos_array.mean()
# È calcolata sull'interezza dell'array, ignorando la sua estensione spaziale
```

```python
# Se vogliamo la media sulle colonne bisogna specificare l'asse
videos_array.mean(axis = 0)
```

```python
videos_array.mean(axis = 1)
```

```python
# Analogamente
videos_array.std(axis = 0)
```

```python
# Per la mediana serve numpy

np.median(videos_array, axis = 0)
```

```python
# Per vedere la correlazione fra i dati numerici

videos.corr()
```

```python
# Per vedere meglio le correlazioni facciamo dei grafici

pl.imshow(videos[ncols[1:]].corr(), clim=(-1,1), cmap="RdBu");
pl.colorbar();
pl.xticks([0, 1, 2, 3], videos[ncols[1:]].columns, rotation=45);
pl.yticks([0, 1, 2, 3], videos[ncols[1:]].columns, rotation=45);
```
