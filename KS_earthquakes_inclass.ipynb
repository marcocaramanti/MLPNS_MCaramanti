```python
# A noi interessa solo la data e la magnitudine
# Trasformiamo data e magnitudine in un data type nuovo
# Prima però diamo alle colonne nomi più comodi.
# Lo facciamo introducendo un dizionario dove
# key: nome attuale
# value: nome nuovo
# Usiamo inplae=True per applicare le modifiche al dataframe che già abbiamo

eqs.rename({"#YYY/MM/DD": "date",
            "HH:mm:SS.ss": "time"}, axis=1, inplace=True)
```

```python
# Questo è il metodo più efficiente per farlo, però vogliamo mettere assieme 
# date e time. Farlo con map produce errore perché alcune righe scazzano.
# Allora ci pieghiamo a farlo con un for

#for i in range(len(eqs)):
#  pd.to_datetime(eqs.iloc[i]["date"] + " " + eqs.iloc[i]["time"])

# Facendo così però otteniamo errore perché c'è un dato con secondi = 60, mentre
# al massimo i secondi arrivano a 59.
# Allora rimpiazziamo 60.00 con 59.99: è un errore di 1 centesimo che non ha
# influenza sulla nostra analisi dato che siamo su timescales ben più lunghe

for i in range(len(eqs)):
  if eqs.iloc[i]["time"].endswith("60.00"):
    eqs.at[i, "time"] = eqs.iloc[i]["time"].replace("60.00", "59.99")
  pd.to_datetime(eqs.iloc[i]["date"] + " " + eqs.iloc[i]["time"])
```
